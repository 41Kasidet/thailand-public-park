```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from pandas import DataFrame
```

# Acquire Link

```python
page_url = 'https://101pub.org/author/kasidet/'

page_source = requests.get(page_url, timeout=10)
page_source.encoding='utf-8'
page_source = page_source.text

page_soup = BeautifulSoup(page_source,'html.parser')
print(page_soup.prettify())
```

```python
# Link อยู่ Tag a in Attribute

# ดึงข้อมูลในส่วนที่เก็บข้อมูลบทความทั้งหมดออกมา
program_container = page_soup.find('div', class_="grid grid-col-1 md:grid-cols-3 gap-8")
# เลือกเฉพาะส่วนที่เก็บข้อมูลหนังสือรายเล่ม Output ที่ได้จะเป็น List รายละเอียดของหนังสือแต่ละเล่ม
program_dat = program_container.find_all('div', class_="border border-gray-300")
```

```python
# Get Link For loop
kasidet_list = []

for a in program_dat:
  # Get Link
  link = a.find('a').attrs['href']
  # Get Title
  title = a.find('h3').text.strip()
  # Get Excerpt
  excerpt = a.find('div', class_="mt-2 text-sm leading-6").text.strip()
  # Append to list
  kasidet_list.append((link,title,excerpt))

df = DataFrame(kasidet_list, columns = ['link','title','excerpt'])
df
```

Adopt Function to Every Page

```python
def page_to_data(page_url):
  page_soup = BeautifulSoup(page_url,'html.parser')
  program_container = page_soup.find('div', class_="grid grid-col-1 md:grid-cols-3 gap-8")
  program_dat = program_container.find_all('div', class_="border border-gray-300")
  for a in program_dat:
    # Get Link
    link = a.find('a').attrs['href']
    # Get Title
    title = a.find('h3').text.strip()
    # Get Excerpt
    excerpt = a.find('div', class_="mt-2 text-sm leading-6").text.strip()
    # Append to list
    kasidet_list.append((link,title,excerpt))
  print('Done')
```

```python
# Run While Loop
kasidet_list = []

url = [
    'https://101pub.org/author/kasidet/',
    'https://101pub.org/author/kasidet/page/2/',
    'https://101pub.org/author/kasidet/page/3/',
    'https://101pub.org/author/kasidet/page/4/']

for url in url:
  try:
    page_url = requests.get(url, timeout=10)
    page_url.encoding='utf-8'
    page_url = page_url.text
    page_to_data(page_url)
    print(f'link {url} is Scraped')
  except:
    print('Every link are Scaped')
    break
```

```python
df = DataFrame(kasidet_list,columns = ['link','title','excerpt'])
df
```

# Scrape บทความตาม Link

```python
# ดูตัวอย่าง

url = df['link'][0]

page_source = requests.get(url, timeout=10)
page_source.encoding='utf-8'
page_source = page_source.text

page_soup = BeautifulSoup(page_source,'html.parser')
print(page_soup.prettify())
```

```python
sentence_list = []

for url in df['link']:
  page_source = requests.get(url, timeout=20)
  page_source.encoding='utf-8'
  page_source = page_source.text
  page_soup = BeautifulSoup(page_source,'html.parser')

  page_container = page_soup.find('div', class_="py-8 space-y-8")
  page_text = page_container.find_all('p')
  list_word = []

  for i in page_text:
    # ต้อง Acquire Text ที่อยู่ 'เฉพาะ Tag 'p' ไม่เอา Text ที่อยู่ในลูกของ Tag 'p' เนื่องจากเป็นการใส่ Endnote ในบทความ ไม่ควรถูกนำมาคำนวณ
    word = i.find_all(string=True, recursive=False)
    # ต้อง Run loop อีกรอบเพื่อใช้ปัญหาที่เกิดจาก Function Fine_all ที่ทำให้ข้อมูลเป็น list
    for j in word:
      word = list_word.append(j)

  sentence_all = ""

  for i in list_word:
    sentence_all = sentence_all+" "+i

  sentence_list.append(sentence_all)
  print(f"{url} is Scraped")

print('Done')
```
